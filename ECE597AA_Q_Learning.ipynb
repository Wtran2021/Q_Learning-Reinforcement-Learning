```
# Imports
import random
from operator import attrgetter
import numpy as np
```

```
# slice class
class Slice:                    # Use this Slice class to store info relevant to each slice
    def __init__(self, type, reward):
        self.type = type;       # Type of slice this is, either 1 2 3 or -1 if none
        self.reward = reward
```

```
# ----network functions----
def assignSlice(kind, reward):  # Create a new slice, include the class of the slice as "kind" and its reward
    newSlice = Slice(kind, reward)
    return newSlice             # Return the new Slice object
```

```
# ----vars----
learnRate = .2                  # learning rate
disRateG = .1                   # gamma
exploration_rate = 1            # exploration rate
max_exploration_rate = 1        # Max Exploration Rate
min_exploration_rate = .01      # Min Exploration Rate
exploration_decay_rate = .05    # Exploration decay rate

count = 1                       # Minute counter, init to 1 so mod works correctly below
hourlyRewards = []              # save hourly net reward to show convergence
numOfHours = 50                  # Num of hours to run the sim for
sliceQueue = []                 # The queue of currently accepted slices, 3 pop'd / hr
q_table = []                    # declare q_table for use later, takes 2 params, curr. state and next action
#q_table = np.zeros((len(sliceQueue), len([0,1])))
```

```
# ----exec----
for hour in range(0, numOfHours):   # Executes hours [0, numOfHours)
    reward = 0                  # hourly reward
    for minute in range(0, 60): # Execute minutes [0, 59)
        possible_slices_list = []
        action = 0 #reset to 0?

        if (count % 5) == 0 and len(sliceQueue) < 4:  # Slice from Class 1 (Utilities)
            possible_slice1 = assignSlice(1, 1)
            # sliceQueue.append(assignSlice(1))  # 1 bc this is class 1
            # reward += 1
            possible_slices_list.append(possible_slice1)

        if (count % 6) == 0 and len(sliceQueue) < 4:  # Slice from Class 3 (Manufacturing)
            possible_slice3 = assignSlice(3, 4)
            # sliceQueue.append(assignSlice(3))
            # reward += 4
            possible_slices_list.append(possible_slice3)

        if (count % 7) == 0 and len(sliceQueue) < 4:  # Slice from Class 2 (Automotive)
            possible_slice2 = assignSlice(2, 2)
            # sliceQueue.append(assignSlice(2))
            # reward += 2
            possible_slices_list.append(possible_slice2)

        if len(possible_slices_list) > 0:       # Only execute below if there is a choice that could be made
            # Q Decision Making
                # either be greedy or wait for a class 3....
            exploration_rate_threshold = random.uniform(0,1)
            if exploration_rate_threshold > exploration_rate:
            #if True:
                # exploit, choose the highest reward possible at this time, or wait(?)
                #action = np.argmax(q_table[sliceQueue, :])
                #Psuedo Q...
                action = max(possible_slices_list, key=attrgetter('reward'))  # this gives us the actual slice
                if len(sliceQueue) < 2 and exploration_rate > 2 * min_exploration_rate: # If there are 1 or fewer items in slice queue, action = whatever is max
                    temp = 1# dont do anything, youre chillin
                elif 2 <= len(sliceQueue) < 3: # if there are =2 or > items, only fill slices of type 2 or 3
                    if action.reward < 2:
                        action.reward = -1 # dont add a slice to sliceQ unless reward is >= 2
                else: # len(sliceQueue) == 3 ...
                    if action.reward < 4:
                        action.reward = -1  # dont add a slice to Q unless its a slice 3 request
                # action = np.argmax(q_table[sliceQueue][:])
                # greedy approach below, choose the available slice with highest reward (commend out when using Q)
                # action = max(possible_slices_list, key=attrgetter('reward'))    # this gives us the actual slice
            else:
                # explore, sample randomly
                action = random.choice(possible_slices_list)                    # this gives us the actual slice

            if action.reward > 0:
                sliceQueue.append(action)  # this appends the choice slice for this minute to our list
                reward += action.reward  # update the hourly reward to reflect accepted slice
            # Update Q table for current action...
            #q_table[state, action]
            #q_table[sliceQueue, action] = q_table[sliceQueue, action] * (1 - learnRate) + learnRate * (action.reward + disRateG * np.max(q_table[(sliceQueue.append(action)), :]))
            #q_table[reward][action] = q_table[reward][action] * (1 - learnRate) + learnRate * (action.reward + disRateG * np.max(q_table[(sliceQueue.append(action)), :]))

        print("Hour %d Min: %d\t[ " % (hour, minute), end='')
        for slices in sliceQueue:
            print("%d " % slices.type, end='')
        print("]\n")
        count += 1
    # End of Hour Updates
    hourlyRewards.append(reward)
    # Updating the exploration chance in the decision making
    exploration_rate = min_exploration_rate + \
                       (max_exploration_rate - min_exploration_rate) * np.exp(-exploration_decay_rate*hour)
    print("----Hour %d Reward = %d----\n\n" % (hour, reward))

    # Finish 3 Slice Requests per hour, so pop them out of the queue
    #print("\n%d" % count)
    sliceQueue.pop(0)
    sliceQueue.pop(0)
    sliceQueue.pop(0)

# End of script Updates
print(hourlyRewards)
print(exploration_rate)
```
